**1. Crea una función que tome un dataframe como parámetro de entrada.** 

La función debe crear (y exportar) un gráfico automáticamente para cada columna del dataframe. Por ejemplo:

> Un histograma/boxplot si la variable es numérica.

> Unas barras de los valores más frecuentes si es categórica.

> Unas barras de los años más frecuentes si el dato está en formato fecha.

La idea es crear una función que funcione para cualquier dataframe, no solo con el que hemos trabajado hasta ahora.

· Muestra el resultado de la función en alguno de los datasets de ejemplo que contiene el paquete seaborn. Por ejemplo: iris, penguins o titanic.

· Ten en consideración que en el siguiente sprint trabajarás exclusivamente con gráficos. El objetivo de este ejercicio no es crear gráficos muy elaborados, sino resolver una necesidad de manera rápida y automática.


```python
import pandas as pd               
import matplotlib.pyplot as plt   # Para crear y guardar los gráficos
import seaborn as sns             # Solo lo usamos para cargar el dataset de ejemplo (titanic)
import os                         # Para crear carpetas en nuestro ordenador


def generar_graficos(dataframe, carpeta_salida='graficos', nombre_dataset='dataset'):
    """
    Parámetros:
        dataframe      : el DataFrame que queremos analizar
        carpeta_salida : carpeta donde se guardarán los gráficos
        nombre_dataset : nombre que aparece en el título de cada gráfico
    """

    # Creamos la carpeta donde se guardarán los gráficos
    # exist_ok=True significa que no dará error si la carpeta ya existe
    os.makedirs(carpeta_salida, exist_ok=True)

    # Recorremos cada columna del DataFrame una por una
    for nombre_columna in dataframe.columns:

        # Guardamos la columna actual en una variable para trabajar con ella más fácil
        columna = dataframe[nombre_columna]

        # CASO 1: La columna es NUMÉRICA (pero no booleana True/False)
        
        # Comprobamos si la columna contiene números
        es_numerica = pd.api.types.is_numeric_dtype(columna)

        # Comprobamos si la columna contiene valores True/False
        # (en pandas los booleanos también se consideran numéricos, por eso los excluimos)
        es_booleana = pd.api.types.is_bool_dtype(columna)

        if es_numerica and not es_booleana:

            # Eliminamos los valores vacíos (NaN) y convertimos todo a número decimal
            datos = columna.dropna().astype(float)

            # Creamos una figura con DOS gráficos apilados verticalmente:
            #   - fila 1: histograma (ocupa 4 partes de altura)
            #   - fila 2: boxplot    (ocupa 1 parte de altura)
            # sharex=True hace que ambos gráficos compartan el mismo eje X
            figura, (eje_histograma, eje_boxplot) = plt.subplots(
                nrows=2,                                   # 2 filas
                ncols=1,                                   # 1 columna
                figsize=(7, 5),                            # tamaño de la figura en pulgadas
                gridspec_kw={'height_ratios': [4, 1]},    # proporción de alturas
                sharex=True                                # comparten el eje X
            )

            # Añadimos un título general a la figura
            figura.suptitle(f'{nombre_dataset}  ·  {nombre_columna}', fontsize=12, fontweight='bold')

            # Dibujamos el histograma en el eje superior
            eje_histograma.hist(
                datos,
                bins=20,           # número de barras del histograma
                color='steelblue', # color de las barras
                edgecolor='white', # color del borde de cada barra
                alpha=0.85         # transparencia (1 = opaco, 0 = invisible)
            )
            # Etiqueta del eje Y del histograma
            eje_histograma.set_ylabel('Frecuencia')

            # Dibujamos el boxplot en el eje inferior
            eje_boxplot.boxplot(
                datos,
                vert=False,        # horizontal (False) en lugar de vertical (True)
                patch_artist=True, # permite colorear el interior de la caja
                boxprops=dict(facecolor='steelblue', alpha=0.6),  # color interior de la caja
                medianprops=dict(color='navy', linewidth=2)        # color y grosor de la mediana
            )
            # Etiqueta del eje X del boxplot (nombre de la columna)
            eje_boxplot.set_xlabel(nombre_columna)

            # Ocultamos las marcas del eje Y del boxplot (no aportan información)
            eje_boxplot.set_yticks([])

        # CASO 2: La columna es de tipo FECHA
        elif pd.api.types.is_datetime64_any_dtype(columna):

            # Extraemos solo el AÑO de cada fecha
            # value_counts() cuenta cuántas veces aparece cada año
            # sort_index() los ordena de menor a mayor
            conteo_años = columna.dropna().dt.year.value_counts().sort_index()

            # Creamos una figura con un solo gráfico
            figura, eje = plt.subplots(figsize=(7, 4))

            # Título de la figura
            figura.suptitle(f'{nombre_dataset}  ·  {nombre_columna}', fontsize=12, fontweight='bold')

            # Dibujamos las barras: una barra por cada año
            # astype(str) convierte los años a texto para que aparezcan bien en el eje X
            eje.bar(conteo_años.index.astype(str), conteo_años.values, color='teal', edgecolor='white')

            # Etiquetas de los ejes
            eje.set_xlabel('Año')
            eje.set_ylabel('Frecuencia')

            # Rotamos las etiquetas del eje X 45 grados para que no se solapen
            eje.tick_params(axis='x', rotation=45)

        # CASO 3: La columna es CATEGÓRICA, BOOLEANA o TEXTO
        else:

            # Cogemos los 10 valores más frecuentes de la columna
            # dropna() elimina los vacíos
            # astype(str) convierte todo a texto (por si hay booleanos)
            # value_counts() cuenta cuántas veces aparece cada valor
            # head(10) se queda solo con los 10 primeros
            valores_frecuentes = columna.dropna().astype(str).value_counts().head(10)

            # Creamos una figura con un solo gráfico
            figura, eje = plt.subplots(figsize=(7, 4))

            # Título de la figura
            figura.suptitle(f'{nombre_dataset}  ·  {nombre_columna}', fontsize=12, fontweight='bold')

            # Dibujamos barras horizontales (barh)
            # [::-1] invierte el orden para que el valor más frecuente quede arriba
            eje.barh(
                valores_frecuentes.index[::-1],   # nombres de los valores (eje Y)
                valores_frecuentes.values[::-1],  # frecuencias (eje X)
                color='coral',
                edgecolor='white'
            )

            # Etiquetas de los ejes
            eje.set_xlabel('Frecuencia')
            eje.set_ylabel(nombre_columna)

        # GUARDADO DEL GRÁFICO (común para los 3 casos)
        # Ajusta automáticamente los márgenes para que no se corten los textos
        plt.tight_layout()

        # Construimos la ruta completa del archivo: carpeta + nombre de columna + .png
        ruta_archivo = os.path.join(carpeta_salida, f'{nombre_columna}.png')

        # Guardamos el gráfico como imagen PNG con resolución de 300 puntos por pulgada
        plt.savefig(ruta_archivo, dpi=300, bbox_inches='tight')

        # Cerramos la figura para liberar memoria (importante en bucles largos)
        plt.close()

        # Mostramos en consola que el gráfico se ha generado correctamente
        print(f'  ✓  {nombre_columna}  →  {ruta_archivo}')

    # Mensaje final con el total de gráficos generados
    print(f'\nSe han generado gráficos para {len(dataframe.columns)} columnas en la carpeta "{carpeta_salida}/"')


# ── EJEMPLO DE USO con el dataset "titanic" de seaborn ───────────────────────

# El bloque if __name__ == '__main__' asegura que este código solo se ejecuta
# cuando lanzamos este archivo directamente, no si lo importamos desde otro script
if __name__ == '__main__':

    # Cargamos el dataset de ejemplo "titanic" que viene incluido en seaborn
    df_titanic = sns.load_dataset('titanic')

    # Mostramos en consola el tamaño del dataset
    print(f'Dataset cargado: titanic  |  {df_titanic.shape[0]} filas × {df_titanic.shape[1]} columnas\n')

    # Llamamos a la función con el dataset del titanic
    generar_graficos(df_titanic, carpeta_salida='graficos_titanic', nombre_dataset='Titanic')


    # ── Para usar con otro propio DataFrame, descomenta estas líneas: ──
    # df_propio = pd.read_csv('mi_fichero.csv')   # cargamos nuestro propio archivo CSV
    # generar_graficos(df_propio, carpeta_salida='mis_graficos', nombre_dataset='Mi Dataset')
```

**2. Carga el archivo matriu_distancies.xlsx en pandas**

De manera que los nombres de filas y los nombres de columnas sean los de las ciudades. Borra "Las Palmas de Gran Canaria" y "Palma" para que podamos hacer el trayecto en coche. (Fuente: Mejores Rutas)

Nos interesa visitar todas las ciudades principales de España recorriendo la mínima distancia posible. No es necesario que lo hagas de forma óptima, nos interesa que desarrolles una solución razonable utilizando las herramientas que tienes actualmente. Por ejemplo, una aproximación sencilla (que no óptima) sería ir siempre a la ciudad más cercana que no hayamos visitado todavía

· Haz una función que, dada la matriz de distancias y la ciudad de origen, haga una propuesta de ruta que sea lo más corta posible que puedas, devolviendo una lista con el orden de visita. Ofrece también la distancia total recorrida.

· (EXTRA): ¿Desde qué ciudad la ruta sería más corta con el algoritmo planteado?

```python
import pandas as pd

# 1. Carga y limpieza 

df = pd.read_excel("matriu_distancies.xlsx", index_col=0)
df.index.name = None  # nombres de fila = ciudades

# Eliminar ciudades no accesibles en coche
df = df.drop(index=["Las Palmas de Gran Canaria", "Palma"],
             columns=["Las Palmas de Gran Canaria", "Palma"])

print("Ciudades cargadas:", df.shape[0])
print(df.index.tolist())

# 2. Algoritmo: vecino más cercano  

def ruta_vecino_cercano(distancias: pd.DataFrame, origen: str) -> tuple[list[str], float]:
    """
    Devuelve (lista_ciudades_en_orden, distancia_total).
    Estrategia: en cada paso ir a la ciudad no visitada más cercana.
    """
    ciudades = distancias.index.tolist()
    visitadas = [origen]
    no_visitadas = set(ciudades) - {origen}
    distancia_total = 0.0

    actual = origen
    while no_visitadas:
        # distancias desde la ciudad actual a todas las no visitadas
        candidatas = distancias.loc[actual, list(no_visitadas)]
        siguiente = candidatas.idxmin()
        distancia_total += candidatas[siguiente]
        visitadas.append(siguiente)
        no_visitadas.remove(siguiente)
        actual = siguiente

    return visitadas, distancia_total


# 3. Ruta desde Sevilla 

origen = "Sevilla"
ruta, total = ruta_vecino_cercano(df, origen)

print(f"\n{'='*55}")
print(f"Ruta desde {origen}")
print(f"{'='*55}")
for i, ciudad in enumerate(ruta, 1):
    print(f"  {i:>2}. {ciudad}")
print(f"\nDistancia total: {total:,.0f} km")

# 4. EXTRA: ¿desde qué ciudad la ruta es más corta? 

print(f"\n{'='*55}")
print("EXTRA – Mejor ciudad de origen")
print(f"{'='*55}")

resultados = {}
for ciudad in df.index:
    _, dist = ruta_vecino_cercano(df, ciudad)
    resultados[ciudad] = dist

mejor_ciudad = min(resultados, key=resultados.get)
mejor_distancia = resultados[mejor_ciudad]

# Top 5
ranking = sorted(resultados.items(), key=lambda x: x[1])
print("\nTop 5 ciudades de origen con menor distancia total:")
for pos, (ciudad, dist) in enumerate(ranking[:5], 1):
    print(f"  {pos}. {ciudad:<35} {dist:>10,.0f} km")

print(f"\n→ La ruta más corta parte desde: {mejor_ciudad} ({mejor_distancia:,.0f} km)")

# Mostrar esa ruta óptima
ruta_opt, _ = ruta_vecino_cercano(df, mejor_ciudad)
print(f"\nRuta completa desde {mejor_ciudad}:")
for i, ciudad in enumerate(ruta_opt, 1):
    print(f"  {i:>2}. {ciudad}")
```
