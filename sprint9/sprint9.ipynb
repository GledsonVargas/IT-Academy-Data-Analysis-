### 1. Calculadora del Índice de Masa Corporal (IMC)

Escribe una función que calcule el IMC introducido por el usuario/a.  
La función debe clasificar el resultado en sus respectivas categorías.

> **Consejo:** Valida los datos previamente para enviar un mensaje de advertencia si los datos introducidos no están en el formato adecuado o no toman valores razonables.

```python
# función para devuelve la fórmula
def calculo_imc(peso, altura):
    return peso / (altura ** 2)

# Pedir al usuario los datos y tratamiento de errores 
def pedir_peso():

    # Validar la variable peso
    while True:
        try:
            peso = float(input("Ingrese su peso en kilos, máximo 250:"))
            if peso <= 0 or peso > 250:
                print("Error: Debes introducir un número razonable. Un número positivo y no superior a 250.")
                continue
            return peso
        except ValueError:
            print("Error: Debes introducir un número positivo y no superior 250.")

  
def pedir_altura():
    
    # Validar la variable altura
    while True:
        try:
            altura = float(input("ingrese su altura en metros (por ejemplo 1.75):"))
            if altura <= 0 or altura > 2.5:
                print("Error: Debes introducir una altura razonable. Máximo 2.50")
                continue
            return altura   
        except ValueError:
            print("Error: Debes introducir un número separado por punto, por ejemplo: 1.75")

# Función para realizar las comparaciones e imprimir el output
def imc():
    print("### BIENVENIDA A LA CALCULADORA DE IMC ###")
    peso = pedir_peso()
    altura = pedir_altura()
    resultado = calculo_imc(peso, altura)
    

    if resultado < 18.5:
        print(f"El resultado es: {round(resultado,2)}. Está bajo su peso, procure engordar de manera saludable.")
    elif resultado <25:
        print(f"El resultado es: {round(resultado,2)}. Está con un peso normal para sus medidas. Mantenerse así.")
    elif resultado <30:
        print(f"El resultado es: {round(resultado,2)}. Se encuentra con sobrepeso. Perder peso es su prioridad.")
    else:
        print(f"El resultado es: {round(resultado,2)}. Pare lo que esté haciendo y busque un médico muchacho. Se encuentra en una situación de obesidad.")

imc()
```

### 3. Contador de palabras de un texto

Escribe una función que, dado un texto, muestre las veces que aparece cada palabra. Intenta gestionar todas las casuísticas posibles que hagan que el programa no funcione correctamente.

> **Extra:** ¿Cuál es la longitud media de las palabras del texto que has escrito? "Hola cómo va?" debería devolver (4+4+2) / 3 = 3,33.

```python
#Recibe los datos del usuario
def frase():
    
    # Verificación de datos
    while True:
        # Variable bandera
        bandera = True
        
        palabras = input("### BIENVENIDO AL CONTADOR DE PALABRAS ### Ingrese una frase (números no serán aceptados):").lower()
        
        # Verificamos cada palabra después de un split, de lo contrario cuenta las palabras
        for n in palabras.split():

            # Limpiamos los signos de puntuación de la palabra antes de verificarla
            palabra_limpia = n.strip(".,;:!¡¿?")
            
            # Si la información no es string
            if palabra_limpia.isalpha() is False:
                # Transaforma la bandera en falsa
                bandera = False
                print("Por favor, ingresar solamente palabras, números y signos no serán contados")
        #Si la bandera es falsa, volvemos el bucle, de lo contrario, return palabras
        if not bandera:
            continue
        else:
            return palabras

palabras = frase()
        
def contador_de_palabras(palabras):
    # Diccionario porque necesitamos dos datos: palabra y un número
    dic_contador = {}
    
    # Verificamos si la palabra está o no en la lista y gestionamos el contador
    for n in palabras.split():

        # limpiamos la palabra antes de meterla al diccionario
        palabra_limpia = n.strip(".,;:!¡¿?")

        if palabra_limpia in dic_contador:
            dic_contador[palabra_limpia] += 1
        else:
            dic_contador[palabra_limpia] = 1
    return dic_contador

def media_palabra(palabras):
    # Contador para sumar las letras sin espacios (por eso el .split)
    contador = 0
    
    for n in palabras.split():
        # limpiamos la palabra antes de contarlas
        palabra_limpia = n.strip(".,;:!¡¿?")
        contador += len(palabra_limpia)
    # media
    media = contador / len(palabras.split())
    return media

# Guardamos los resultados e imprimimos en pantalla
resultado = contador_de_palabras(palabras)
for clave, valor in resultado.items():
    print(f"{clave}: {valor}")

resultado_media = media_palabra(palabras)
print(f"La media de letras en las palabras en su frase es {round(resultado_media,2)}")
```

### 4. Diccionario Inverso (con posibilidad de duplicados)

Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y necesita los resultados al revés, es decir, intercambiando las claves y los valores. Los valores y claves en el diccionario original son únicos; si este no es el caso, la función debería imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.

> ** Fíjate en los siguientes ejemplos para ver el output esperado. (imágenes de referencia en el enunciado original) **



```python
dic_normal = {
    "a":1, 
    "b":2,
    "c":"d",
    "z":"d",
    "f":"d",
    "e":2
}
def dic_invertido(dict):
    
    # Guardamos el resultado vacío
    reverse_dictionary = {}
    
    for clave, valor in dict.items():
        # El primer filtro: ¿Ya existe eso?
        # Si la respuesta es NO, va directamente al else
        # lo guarda de forma normal pero invertido
        if valor in reverse_dictionary:
            
            # Si la respuesta es SÍ
            # isinstance verifica si es una lista y por lo tanto el append
            if isinstance(reverse_dictionary[valor], list):
                reverse_dictionary[valor].append(clave)
            # Si no es una lista (una letra suelta por ejemplo)
            # Transformamos en una lista combinando lo que ya había con la nueva letra
            else:
                reverse_dictionary[valor] = [reverse_dictionary[valor], clave]
        else:
            reverse_dictionary[valor] = clave
    return reverse_dictionary

resultado = dic_invertido(dic_normal)
print(resultado)
```

**PASO A PASO**

Vuelta 3 ("c": "d"): El valor "d" no existe en el nuevo diccionario. Se va al else y guarda: {"d": "c"}.

Vuelta 4 ("z": "d"): El valor "d" ya existe. Python pregunta: ¿"c" es una lista? No, es una letra suelta. Entonces ejecuta el else interno y lo transforma en lista: {"d": ["c", "z"]}.

Vuelta 5 ("f": "d"): El valor "d" ya existe. Python pregunta: ¿["c", "z"] es una lista? ¡Sí! Entonces ejecuta el if interno y simplemente añade "f" al final: {"d": ["c", "z", "f"]}.

### 1. Contador y ordenador de palabras de un texto

El cliente quedó contento con el contador de palabras, pero ahora quiere leer archivos TXT y que calcule la frecuencia de cada palabra ordenadas dentro de las entradas habituales del diccionario según la letra con la que empiezan, es decir, las claves deben ir de la A a la Z y dentro de cada letra, también de la A a la Z. Por ejemplo, para el archivo "tu_me_quieres_blanca.txt" la salida esperada sería la que se muestra en la imagen de referencia.



```python
# Leer todo el contenido como string y cerrarlo después
with open("tu me quieres blanca.txt", encoding="utf-8") as f:
    contenido = f.read()

def contador_de_textos(text):

    diccionario = {}
    
    for n in text.lower().split(): # > .lower para que minúscula y mayúscula cuente como misma palabra

        n = n.strip("!,().¡") # > Quita los signos de puntuación que estén pegados a las palabras.
        
        # If n es FALSE, es decir
        # Si por alguna razón al limpiar un signo nos quedamos con una palabra vacía, CONTINUE le dice a Python: "Salta esta vuelta y ve a la siguiente palabra, aquí no hay nada que ver"
        if not n: 
            continue
        
        # Guardamos la primera letra de la palabra
        letra = n[0]

        # Python pregunta: ¿Ya creé un cajón para la letra "b" de Blanca, por ejemplo?
        if letra in diccionario:

            # ¿Esta palabra exacta (ej. "blanca") ya la había anotado dentro del cajón de la "b"?
            if n in diccionario[letra]:
                diccionario[letra][n] += 1 # Sí. Le suma uno al contador. Ya van 2 veces que aparece "blanca"
            else:
                diccionario[letra][n] = 1 # No. Es una palabra nueva para ese cajón, por ejemplo "buena", así que la anota por primera vez con un 1.      
        
        # Como es la primera vez que ve una palabra con esa letra (la "b" de "blanca"):
        else:
            diccionario[letra] = {} # Crea un cajón vacío exclusivo para la letra "b"
            diccionario[letra][n] = 1 # Mete la palabra "blanca" en ese cajón y le pone el contador en 1.
    
    # sorted(diccionario.items()): Ordena los "cajones" alfabéticamente (la a primero, luego la b, luego la c...).
    # sorted(palabras.items()): Ordena las palabras que están dentro de cada cajón también alfabéticamente
    return {letra: dict(sorted(palabras.items())) for letra, palabras in sorted(diccionario.items())}

print(contador_de_textos(contenido))

```

### 2. Conversión de tipos de datos

El cliente recibe una lista de datos y necesita generar dos listas: la primera con todos los elementos que se pudieron convertir a flotantes, y la otra con los elementos que no se pudieron convertir. Ejemplo de la lista que recibe el cliente:

> ** ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, '3', '3.4']] **



Pseudocódigo:

- **Verificar cada elemento**
- **Es lista o tupla? → bajar un nivel y repetir el proceso con cada elemento interno**
- **No lo es? → intentar convertir a float normalmente**


```python
# Valores a verificar
lista = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

# Función para clasificar entre float/no float
def clasificar(lista):

    # Listas vacías para acomodar los floats/no floats
    floats = []
    no_floats = []

    # Para cada elemento en la lista
    for n in lista:
        # Primero verifica si es una lista o tupla
        if isinstance (n, (tuple, list)):
            # Analiza cada elemento dentro de la lista/tupla y lo transforme en float si posible
            # Como n ahora no representa un elemento sino un lista/tupla
            # El "elemento" funciona como "n"
            for elemento in n:
                try:
                    floats.append(float(elemento))
                # Si no se puede, agrégalo a la lista de los que no
                except(ValueError, TypeError):
                    no_floats.append(elemento)
        
        # Ese else es para los elementos sueltos, que no entran en el primer filtro
        else:
            try:
                floats.append(float(n))
            except (ValueError, TypeError):
                no_floats.append(n)

    return floats, no_floats 
            

resultado = clasificar(lista)

print(resultado)
```

### 1. Generador de contraseñas

· Explora el funcionament del mòdul random de la llibreria numpy.  
· En este punto, el cliente ha detectado un problema con las contraseñas que utilizan sus trabajadores: Asdf1234, fechas de cumpleaños o similares. Para solucionarlo nos ha encargado una función de Python que genere contraseñas más seguras. La función debe depender de los siguientes parámetros:

> longitud (int): Longitud de la contrasenya

> mayusculas (bool = True): Si deben aparecer mayúsculas

> minusculas (bool = True): Si deben aparecer minúsculas

> numeros (bool = True): Si deben aparecer números

> signos (bool = False): Si deben aparecer caracteres especiales (,-$? o similares)

Así pues, si ejecutamos la función de la siguiente manera:

**· crear_contraseña(10, True, True, True, True)**

Deberíamos obtener un output (que debemos poder guardar) del estilo:
>**9Er,5Vn8P$**

Asegúrate de que se cumplan todos los criterios y de que estas contraseñas sean realmente aleatorias.

> (EXTRA) Explora cómo podríamos hacer que la función copiara la contraseña automáticamente al portapapeles del ordenador (como si la hubiésemos seleccionado y pulsado Ctrl+C).

 construir pool > garantizar mínimos > rellenar aleatoriamente > barajar.

```python
import numpy as np
import string
import pyperclip
from numpy import random

# Esa linea el lo último que aprendí. Yo no veo donde está obligado a que tenga o no los 4 tipos
# Eso puede ser definido a la hora de llamar la función, con los True y False
def crear_contraseña(longitud, mayusculas=True, minusculas=True, numeros=True, signos=False):

    # Lista para recibir el número random
    pools_activos = []
    #Lista auxiliar para verificar si hay un número de cada tipo
    pool_obligatorio = []

    # Defino pool de caracteres a incluir en la contraseña    
    if mayusculas:
        pools_activos += list(string.ascii_uppercase) # Si son mayúsculas, añadirlas a la lista, lo mismo con minúsculas, número y simbolos
        pool_obligatorio.append(random.choice(list(string.ascii_uppercase))) # Añadir a la lista auxiliar y también transformar el string en lista porque acepta strings
    if minusculas:
        pools_activos += list(string.ascii_lowercase)
        pool_obligatorio.append(random.choice(list(string.ascii_lowercase)))
    if numeros:
        pools_activos += list(string.digits)
        pool_obligatorio.append(random.choice(list(string.digits)))
    if signos:
        pools_activos += list(string.punctuation)
        pool_obligatorio.append(random.choice(list(string.punctuation)))

    # random.choice(de dónde elegir, cuantos)
    # Join es para eliminar los espacios. Salida: 'HolamundoenPython'
    # Aquí tengo la longitud menos los obligatorios
    contraseña = "".join(np.random.choice(pools_activos, longitud - len(pool_obligatorio)))

    # Justamos los dos para sumar 10 con 1 de cada obligatorio
    contraseña_final = pool_obligatorio + list(contraseña)

    # Hacemos un shuffle para randomizarlo todo
    random.shuffle(contraseña_final)

    # Transformamos en string todo junto
    contraseña_final = "".join(contraseña_final)
    return contraseña_final

# Ejemplo de uso
resultado = crear_contraseña(10, True, True, True, True)
print(resultado)

# Copiar la contraseña al portapapeles
pyperclip.copy(resultado)
```

· Una compañera de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar. Tiene un histórico de partidos de fútbol catalán en un archivo, donde están almacenados los nombres de los equipos y los resultados. Necesita que procesemos los datos de forma automática para extraer los resultados que necesita.

> Utiliza el archivo "historic_partits.txt".

Necesita un programa que devuelva:

**· El número total de goles que ha marcado cada equipo.**

**· El nombre del equipo más goleador.**

**· El nombre del equipo más goleado.**

**· La clasificación global (cada victoria: 3 pts, empate: 1 pt, derrota: 0 pts).**

Como referencia, los equipos con más puntos serían los que se muestran en la imagen:



Puedes utilizar el formato que te sea más conveniente.

```python
import pandas as pd

with open("historic partits.txt", encoding="utf-8") as f:
    historial_partidos = f.readlines()

def goles_partidos(texto):
    # diccionar
    goles_marcados = {}
    goles_encajados = {}
    puntos = {}

    for linea in historial_partidos:
        partes = linea.strip().split('\t') # -> ["Girona FC", "0-3", "RCD Espanyol"]
        equipo_local = partes[0] # "Girona FC"
        equipo_visitante = partes[2] # "RCD Espanyol"
        goles = partes[1].split("-") # -> Pasamos de ["0-3"] a ["0", "3"]
        goles_local = int(goles[0]) # Pasamos a int para poder hacer cálculos
        goles_visitante = int(goles[1]) 
        
        # Inicializar claves si no existen
        if equipo_local not in goles_marcados:
            goles_marcados[equipo_local] = 0
        if equipo_visitante not in goles_marcados:
            goles_marcados[equipo_visitante] = 0
        if equipo_local not in goles_encajados:
            goles_encajados[equipo_local] = 0
        if equipo_visitante not in goles_encajados:
            goles_encajados[equipo_visitante] = 0
        if equipo_local not in puntos:
            puntos[equipo_local] = 0
        if equipo_visitante not in puntos:
            puntos[equipo_visitante] = 0
        
        # Sumar goles
        goles_marcados[equipo_local] += goles_local
        goles_marcados[equipo_visitante] += goles_visitante
        goles_encajados[equipo_local] += goles_visitante
        goles_encajados[equipo_visitante] += goles_local

        # Calcular puntos
        if goles_local > goles_visitante:
            puntos[equipo_local] += 3
        elif goles_local < goles_visitante:
            puntos[equipo_visitante] += 3
        else:
            puntos[equipo_local] += 1
            puntos[equipo_visitante] += 1

    # Construir DataFrames
    df_clasificacion = pd.DataFrame(puntos.items(), columns=["Equipo", "Puntos"])
    df_clasificacion = df_clasificacion.sort_values("Puntos", ascending=False).reset_index(drop=True)

    df_goles = pd.DataFrame({
        "Equipo": goles_marcados.keys(),
        "Goles marcados": goles_marcados.values(),
        "Goles encajados": goles_encajados.values()
    })

    df_final = df_clasificacion.merge(df_goles, on="Equipo")

    mas_goleador = max(goles_marcados, key=goles_marcados.get)
    mas_goleado = max(goles_encajados, key=goles_encajados.get)

    print(df_final.to_string(index=False))
    print(f"\nEquipo más goleador: {mas_goleador}")
    print(f"Equipo más goleado: {mas_goleado}")

goles_partidos(historial_partidos)
```
